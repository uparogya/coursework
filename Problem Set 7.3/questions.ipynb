```python
import numpy as np
from sympy import Matrix
import matplotlib.pyplot as plt
```

```python
A0 = np.array([
    [5,4,3,2,1],
    [-1,1,0,1,-1]
])
average_of_row_1 = 0

print("A0 is \n",A0)

for element in A0[0]:
    average_of_row_1 += element

average_of_row_1 = average_of_row_1 / len(A0[0])
print("\nDifference of averages is: ", average_of_row_1, "\n")


U, S, VH = np.linalg.svd(A0, full_matrices = True)

print("U is \n", U, "\n")
print("S is \n", S, "\n")
print("VH is \n", VH, "\n")

rank = np.linalg.matrix_rank(A0)
Smat = np.zeros(A0.shape)
Smat[:rank, :rank] = np.diag(S)
print(Smat)

A_recon = U @ Smat @ VH
print("\n The reconstructed A0 matrix is: \n\n", A_recon)
```

```python
A = A0.copy()
A[0, :] = A[0, :] - 3
print(A)

S = (A @ A.T) / (len(A[0, :]))
print("\n", S)

np.linalg.eig(S)
```

```python
plt.scatter(A0[0, :], A[1, :])
plt.show()
```
